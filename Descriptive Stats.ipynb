```python
target = pd.read_csv("Customer_Response_data_P1.csv")
target.head(3)
```

```python
cust_outcome = pd.merge(custDetails, target, on='Customer_id', how='inner')
cust_outcome.reset_index(drop=True, inplace=True)
cust_outcome.tail()
```

```python
analysis_education = pd.pivot_table(cust_outcome, index=['education'], values=['Customer_id'], columns=['y'])
total = analysis_education['Customer_id']['yes']+analysis_education['Customer_id']['no']
analysis_education['SuccessRate']=(analysis_education['Customer_id']['yes']/total)*100
analysis_education = analysis_education.reindex(analysis_education.sort_values(by='SuccessRate', ascending=False).index)
analysis_education
```

The average success rate is highest for customers with education level of 4years courses.
